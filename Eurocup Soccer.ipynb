## Questions

### 1. How many teams participated in the Euro2012?

```python
Teams_participated = df1['Team'].nunique()
print('The Total number of teams participated:',Teams_participated)
```

### 2. What is the number of columns in the dataset?

```python
df1.shape
print('The Total number of column in the dataset:',df1.shape[1])
```

### 3. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline.

```python
discipline = df1[['Team','Yellow Cards','Red Cards']]
discipline
```

### 4. Sort the teams by Red Cards, then to Yellow Cards.

```python
sorted_discipline = discipline.sort_values(by=['Red Cards','Yellow Cards'])
sorted_discipline
```

### 5. Calculate the mean Yellow Cards given per Team.

```python
mean_yellow_cards = discipline['Yellow Cards'].mean()
print('Mean of Yellow cards given per team :',mean_yellow_cards)
```

### 6. Filter teams that scored more than 6 goals.

```python
filtered_df = df1[df1['Goals'] > 6]
filtered_df = filtered_df[['Team','Goals']]
filtered_df
```

### 7. Select the teams that start with the letter G.

```python
G_teams = df1[df1['Team'].str.startswith('G')]
G_teams = G_teams[['Team']]
G_teams
```

### 8. Select the first 7 columns.

```python
first_7_columns = df1.iloc[:, :7]
first_7_columns
```

### 9. Select all columns except the last 3.

```python
except_last_3_columns = df1.iloc[:, :-3]
except_last_3_columns
```

### 10. Present only the Shooting Accuracy from England, Italy and Russia.

```python
Shooting_accuracy = df1[df1['Team'].isin(['England','Italy','Russia'])]
Shooting_accuracy = Shooting_accuracy[['Team','Shooting Accuracy']]
Shooting_accuracy
```
